```python
import pandas as pd

df1 = pd.read_csv("csv_files/file1.csv")
df1
df1.shape

import re
import string
df1.columns = df1.columns.str.lower()
print(df1.columns)
df1
```

```python

df2 = pd.read_csv("csv_files/file2.csv")
df2.shape
df2.columns = df2.columns.str.lower()
print(df2.columns)
df2
```

```python

df3 = pd.read_csv("csv_files/file3.csv")
df3.shape
df3.columns = df3.columns.str.lower()
print(df3.columns)
df3
```

```python
df_all = pd.concat([df1, df2, df3], axis=0)
df_all
df_all = df_all.rename(columns= {"customer lifetime value": "CLV", "monthly premium auto": "MPA", "number of open complaints": "NOC", "policy type": "Policy_type", "vehicle class": "Vehicle_class", "total claim amount": "TCA"})
df_all
```

```python
#numerical columns = Income, monthly premium auto, number of open complanis, costumer lifetime value
##categorical columns = Costumer, ST, Gender, Vehicle class, education
```

```python
df_all.drop(labels=["education", "NOC"], axis=1)

df_all.info()
df_all.CLV.isna().sum()
df_all = df_all.dropna(subset = ["CLV"])
df_all
```

```python
#df_all["CLV"] = pd.to_numeric(df_all['CLV'], errors = 'ignore')
df_all["CLV"] = df_all["CLV"].str.replace("%", "")
df_all["CLV"] = df_all["CLV"].astype(float)
df_all
df_all.info()
```

```python
#df_all[df_all.select_dtypes(include=['CLV'])] *= 100
df_all
df_all.info()
```

```python
df_all["CLV"] = df_all["CLV"].div(1000)
df_all
```

```python
dr = df_all.duplicated()
print(dr)
```

```python
df_all.fillna(value={'st': 'no data', 'CLV': 0, 'state': "no data", "gender": "non binary"}, inplace=True)
df_all
```
